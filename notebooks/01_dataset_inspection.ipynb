```python
import pandas as pd

# Helper function
def extract_value(x):
    if isinstance(x, list) and len(x) > 0:
        first = x[0]
        if isinstance(first, dict):
            return first.get("value")
    return None

# Create cleaned master table
catalog_df = pd.DataFrame({
    "item_id": full_df["item_id"],
    "title": full_df["item_name"].apply(extract_value),
    "product_type": full_df["product_type"].apply(extract_value),
    "brand": full_df["brand"].apply(extract_value),
    "color": full_df["color"].apply(extract_value),
})

# Remove rows with missing title or product type
catalog_df = catalog_df.dropna(subset=["title", "product_type"]).reset_index(drop=True)

print(f"Clean catalog shape: {catalog_df.shape}")
print()
print(catalog_df.head())
```

```python
# Most common product types
top_types = catalog_df["product_type"].value_counts().head(20)

print(top_types)
```

```python
# Search for product types related to bags, watches, and kitchen items

all_types = catalog_df["product_type"].unique()

keywords = ["BAG", "WATCH", "KITCHEN", "COOK", "DRINK", "CUP", "DINING"]

for keyword in keywords:
    matches = [t for t in all_types if keyword in str(t)]
    print(f"\n=== {keyword} ===")
    for m in matches[:20]:
        count = (catalog_df["product_type"] == m).sum()
        print(f"{m:35} {count}")
```

```python
CATEGORY_MAPPING = {
    "Footwear": [
        "SHOES", "BOOT", "SANDAL"
    ],
    
    "Bags": [
        "HANDBAG", "TOTE_BAG", "STORAGE_BAG"
    ],
    
    "Watches": [
        "WATCH", "SWATCH"
    ],
    
    "Furniture": [
        "CHAIR", "SOFA", "TABLE", "HOME_FURNITURE_AND_DECOR"
    ],
    
    "Kitchenware": [
        "KITCHEN", "DRINKING_CUP", "MEASURING_CUP", "BAKING_CUP"
    ],
    
    "Electronics_Accessories": [
        "CELLULAR_PHONE_CASE", "HARDWARE_HANDLE",
        "CAMERA_BAGS_AND_CASES", "CARRYING_CASE_OR_BAG"
    ]
}
```

```python
# Reverse mapping: product_type -> target category
type_to_category = {}

for category, types in CATEGORY_MAPPING.items():
    for t in types:
        type_to_category[t] = category

# Create target category column
catalog_df["target_category"] = catalog_df["product_type"].map(type_to_category)

# Keep only mapped categories
filtered_df = catalog_df.dropna(subset=["target_category"]).reset_index(drop=True)

print("Filtered dataset shape:", filtered_df.shape)
print()
print(filtered_df["target_category"].value_counts())
```

```python
import pandas as pd

# Keep only categories with enough samples
selected_categories = ["Footwear", "Furniture", "Kitchenware", "Bags"]

subset_df = filtered_df[
    filtered_df["target_category"].isin(selected_categories)
].copy()

print("Available samples:")
print(subset_df["target_category"].value_counts())
```

```python
# Target samples per category
TARGET_SAMPLES = {
    "Footwear": 3000,
    "Furniture": 3000,
    "Kitchenware": 1000,
    "Bags": 900
}

balanced_parts = []

for category, n_samples in TARGET_SAMPLES.items():
    category_df = subset_df[
        subset_df["target_category"] == category
    ]
    
    sampled = category_df.sample(
        n=n_samples,
        random_state=42
    )
    
    balanced_parts.append(sampled)
    
    print(f"{category}: {len(sampled)} samples selected")

# Combine and shuffle
balanced_df = pd.concat(balanced_parts, ignore_index=True)
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

print()
print(f"Final balanced dataset shape: {balanced_df.shape}")
print()
print(balanced_df["target_category"].value_counts())
```

```python
# Save processed dataset
output_path = project_root / "data" / "processed" / "catalog_subset.parquet"

balanced_df.to_parquet(output_path, index=False)

print(f"Saved to: {output_path}")
```

```python
import os

file_size_mb = os.path.getsize(output_path) / (1024 * 1024)

print(f"File size: {file_size_mb:.2f} MB")
print()
print(balanced_df.head())
```

```python
from sklearn.model_selection import train_test_split

# First split: Train (70%) and Temp (30%)
train_df, temp_df = train_test_split(
    balanced_df,
    test_size=0.30,
    stratify=balanced_df["target_category"],
    random_state=42
)

# Second split: Validation (15%) and Test (15%)
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    stratify=temp_df["target_category"],
    random_state=42
)

print("Train shape:", train_df.shape)
print("Validation shape:", val_df.shape)
print("Test shape:", test_df.shape)

print("\nTrain distribution:")
print(train_df["target_category"].value_counts())

print("\nValidation distribution:")
print(val_df["target_category"].value_counts())

print("\nTest distribution:")
print(test_df["target_category"].value_counts())
```

```python
# Save splits
splits_dir = project_root / "data" / "splits"

train_path = splits_dir / "train.parquet"
val_path = splits_dir / "val.parquet"
test_path = splits_dir / "test.parquet"

train_df.to_parquet(train_path, index=False)
val_df.to_parquet(val_path, index=False)
test_df.to_parquet(test_path, index=False)

print("Saved files:")
print(train_path)
print(val_path)
print(test_path)
```
